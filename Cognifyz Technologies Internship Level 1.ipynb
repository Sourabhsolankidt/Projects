 Task:  Descriptive Analysis
 Calculate basic statistical measures (mean,
 median, standard deviation, etc.) for numerical
 columns.
 Explore the distribution of categorical
 variables like "Country Code," "City," and
 "Cuisines."
 Identify the top cuisines and cities with the

```python
# Calculate basic statistical measures for numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns
print("Basic Statistical Measures:")
for col in numerical_cols:
    print(f"Column: {col}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")
    print(f"Minimum: {df[col].min():.2f}")
    print(f"Maximum: {df[col].max():.2f}")
    print("")
```

```python
# Explore the distribution of categorical variables
categorical_cols = df.select_dtypes(include=[object]).columns
print("Categorical Variable Distribution:")
for col in categorical_cols:
    print(f"Column: {col}")
    print(df[col].value_counts())
    print("")
```

```python
# Identify the top cuisines and cities with the highest number of restaurants
print("Top Cuisines:")
top_cuisines = df['Cuisines'].value_counts().head(10)
print(top_cuisines)
```

```python
print("Top Cities:")
top_cities = df['City'].value_counts().head(10)
print(top_cities)
```

```python
# Visualize the distribution of categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x="Cuisines", data=df)
plt.title("Distribution of Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="City", data=df)
plt.title("Distribution of Cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr
```

```python
# Create a Folium map to visualize restaurant locations
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=4)
```

```python
# Add markers for each restaurant
marker_cluster = MarkerCluster().add_to(m)
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(marker_cluster)
```

```python
# Save the map to an HTML file
m.save('Dataset_map.html')
```

```python
# Analyze the distribution of restaurants across different cities or countries
print("Distribution of Restaurants across Cities:")
city_counts = df['City'].value_counts()
print(city_counts)
```

```python
print("Distribution of Restaurants across Countries:")
country_counts = df['Country Code'].value_counts()
print(country_counts)
```

```python
# Visualize the distribution of restaurants across cities and countries
plt.figure(figsize=(10, 6))
sns.countplot(x="City", data=df)
plt.title("Distribution of Restaurants across Cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Country Code", data=df)
plt.title("Distribution of Restaurants across Countries")
plt.xlabel("Country Code")
plt.ylabel("Count")
plt.show()
```

```python
latitude_corr, _ = spearmanr(df['Latitude'], df['Rating'].rank())
print(f'Correlation between Latitude and Rating: {latitude_corr:.2f}')
```

```python
# Calculate the correlation between Longitude and Rating
longitude_corr, _ = spearmanr(df['Longitude'], df['Rating'].rank())
print(f'Correlation between Longitude and Rating: {longitude_corr:.2f}')
```

```python
# Group by city and count the number of restaurants
city_counts = df.groupby('City')['Restaurant Name'].count()
# Group by country and count the number of restaurants
country_counts = df.groupby('Country')['Restaurant Name'].count()
print("Distribution of restaurants by country:")
print(country_counts)
print("Distribution of restaurants by city:")
print(city_counts)
```
