# Task 1
 Task: Table Booking and Online Delivery
 Determine the percentage of restaurants that
 offer table booking and online delivery.
 Compare the average ratings of restaurants
 with table booking and those without.
 Analyze the availability of online delivery
 among restaurants with different price ranges

```python
# Calculate the percentage of restaurants that offer table booking
table_booking_yes = df['Has Table booking'].value_counts()['Yes']
table_booking_no = df['Has Table booking'].value_counts()['No']
table_booking_percentage = (table_booking_yes / (table_booking_yes + table_booking_no)) * 100
print(f"Percentage of restaurants that offer table booking: {table_booking_percentage:.2f}%")

# Calculate the percentage of restaurants that offer online delivery
online_delivery_yes = df['Has Online delivery'].value_counts()['Yes']
online_delivery_no = df['Has Online delivery'].value_counts()['No']
online_delivery_percentage = (online_delivery_yes / (online_delivery_yes + online_delivery_no)) * 100
print(f"Percentage of restaurants that offer online delivery: {online_delivery_percentage:.2f}%")
```

```python
# Calculate the average rating of restaurants with table booking
avg_rating_table_booking = df[df['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
print(f"Average rating of restaurants with table booking: {avg_rating_table_booking:.2f}")

# Calculate the average rating of restaurants without table booking
avg_rating_no_table_booking = df[df['Has Table booking'] == 'No']['Aggregate rating'].mean()
print(f"Average rating of restaurants without table booking: {avg_rating_no_table_booking:.2f}")
```

```python
# Analyze the availability of online delivery among restaurants with different price ranges
price_ranges = ['Low', 'Medium', 'High']
online_delivery_by_price_range = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True) * 100
```

```python
print('Has Online delivery availability by price range:')
print(online_delivery_by_price_range)
```

```python
# Group by price range and calculate the percentage of restaurants with online delivery
price_ranges = ['Low', 'Medium', 'High']
for price_range in price_ranges:
    online_delivery_yes = df[(df['Price range'] == price_range) & (df['Has Online delivery'] == 'Yes')].shape[0]
    online_delivery_no = df[(df['Price range'] == price_range) & (df['Has Online delivery'] == 'No')].shape[0]
    
    if online_delivery_yes + online_delivery_no == 0:
        print(f"No restaurants found in {price_range} price range.")
    else:
        online_delivery_percentage = (online_delivery_yes / (online_delivery_yes + online_delivery_no)) * 100
        print(f"Percentage of restaurants with online delivery in {price_range} price range: {online_delivery_percentage:.2f}%")
```

# Task 2
 Task: Price Range Analysis
 Determine the most common price range
 among all the restaurants.
 Calculate the average rating for each price
 range.
 Identify the color that represents the highest
 average rating among different price ranges

```python
    # Determine the most common price range
most_common_price_range = df['Price range'].value_counts().index[0]
print(f"The most common price range is: {most_common_price_range}")
```

```python
# Calculate the average rating for each price range
average_ratings = df.groupby('Price range')['Aggregate rating'].mean()
print("Average ratings for each price range:")
print(average_ratings)
```

```python
# Identify the color that represents the highest average rating
highest_average_rating = average_ratings.max()
color_map = {1: 'Blue', 2: 'Orange', 3: 'Red', 4: 'Teal'}
highest_average_rating_color = color_map[average_ratings.idxmax()]
print(f'Color that represents the highest average rating: {highest_average_rating_color}')
```

# Task 3
 Task: Feature Engineering
 Extract additional features from the existing
 columns, such as the length of the restaurant
 name or address.
 Create new features like "Has Table Booking"
 or "Has Online Delivery" by encoding
 categorical variables

```python
# feature_engineering.py
# Extract additional features
df['Restaurant Name Length'] = df['Restaurant Name'].apply(len)
df['Address Length'] = df['Address'].apply(len)

# Create new features by encoding categorical variables
df['Has Table booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})

# One-Hot Encoding for categorical variables
categorical_cols = ['Country Code', 'City', 'Cuisines']
for col in categorical_cols:
    one_hot = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df, one_hot], axis=1)
    df.drop(col, axis=1, inplace=True)
```
